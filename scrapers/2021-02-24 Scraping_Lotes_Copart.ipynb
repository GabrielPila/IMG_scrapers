```python
%%time
# autos = {}
for index, row in lotes_autos[~(lotes_autos['lote'].isin(autos.keys()))].iterrows():
    
    try:
        features = {}

        lote=row['lote']
        print(index, lote)

        url = f'https://www.copart.com/es/lot/{lote}'
        driver.get(url)
        time.sleep(4)

        driver.find_element_by_css_selector('a.viewalltxt').click()
        time.sleep(2)

        features['imgs'] = [x.find_element_by_css_selector('img').get_attribute('src') for x in driver.find_elements_by_css_selector('div.viewAllPhotosRelative')]
        try: features['VIN'] = driver.find_elements_by_css_selector('div.clearfix')[3].text.split('\n')[1]
        except: features['VIN'] = np.nan
        try: features['odometro'] = driver.find_elements_by_css_selector('div.clearfix')[5].text.split('\n')[1]
        except: features['odometro'] = np.nan
        try: features['danho_principal'] = driver.find_elements_by_css_selector('div.clearfix')[7].text.split('\n')[1]
        except: features['danho_principal'] = np.nan
        features['clase_buscada'] = row['clase_veh']
        features['marca_modelo_buscada'] = row['marca_modelo']
        features['danho_grave'] = row['danho_grave']
        features['lote'] = row['lote']
        autos[lote] = features

        with open(f'autos_lotes_indira.json', 'w') as fp:
            json.dump(autos, fp)
    
    except:
        print(f'{index} - {row["lote"]} no se pudo procesar.')
        pass
```

```python
lista_clases = []
piezas_fotos = {'REMOLQUE': [],'CMTA PICKUP': [],'CMTA RURAL': [],'AUTOMOVIL': [],'CMTA PANEL': [],'CAMION': []}
vin_fotos = {'REMOLQUE': [],'CMTA PICKUP': [],'CMTA RURAL': [],'AUTOMOVIL': [],'CMTA PANEL': [],'CAMION': []}
odometro_fotos = {'REMOLQUE': [],'CMTA PICKUP': [],'CMTA RURAL': [],'AUTOMOVIL': [],'CMTA PANEL': [],'CAMION': []}

dict_fotos = {}
for lote in autos:
    clase_veh = autos[lote]['clase_buscada']
    imgs = autos[lote]['imgs']
    if len(imgs)==10:
        piezas_fotos[clase_veh].append(imgs[:4])
        vin_fotos[clase_veh].append(imgs[-1])
        odometro_fotos[clase_veh].append(imgs[7])
    else:
        pass
dict_fotos['piezas_fotos'] = piezas_fotos
dict_fotos['vin_fotos'] = vin_fotos
dict_fotos['odometro_fotos'] = odometro_fotos
```
